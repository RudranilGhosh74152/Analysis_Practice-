# Find average value of reviews

```python
gplay_data.columns
```

One row in the dataset had 3.0M in Reviews column so changing the value to 3.0 to apply mathematical operations.

```python
gplay_data['Reviews'].replace('3.0M','3.0',inplace=True)
```

Changing datatype of values in Reviews from string to float.

```python
gplay_data.Reviews=gplay_data.Reviews.astype(float)
```

```python
gplay_data.Reviews.dtype
```

```python
print('The average value of reviews:',gplay_data.Reviews.mean())
```
