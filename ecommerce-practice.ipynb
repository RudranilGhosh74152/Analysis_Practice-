# How many people have a credit card that expires in 2020?

```python
data.columns
```

```python
data['CC Exp Date'].dtype
```

```python
data['CC Exp Date'].head()
```

```python
def takeYear(n):
    count =0
    for d in data['CC Exp Date']:
        x=d.split('/')
        if x[1]==n:
                count+=1
    print(count) 
```

```python
takeYear('20')
```

```python
len(data[data['CC Exp Date'].apply(lambda x:x[3:]=='20')])
```
