# Finding average gross income according to cities.

```python
df.columns
```

```python
df.pivot_table(index='City',values='gross income',aggfunc='mean')
```

In Naypyitaw city we have higher gross income.

```python
df.pivot_table(index='City',values='gross income',aggfunc='mean').plot(kind='bar',xlabel='City',ylabel='Gross Income',fontsize=10,figsize=(5,5),color='lightgreen')
plt.xticks(rotation=70 ,fontsize=11)
```

# Finding which product line has maximum gross income.

```python
df.columns
```

```python
df.pivot_table(index='Product line',values='gross income').sort_values(by='gross income',ascending =False)
```

```python
df.pivot_table(index='Product line',values='gross income').plot()
plt.xticks(rotation=60)
plt.show()
```

After visualizing we can conclude Home and lifestyle product line has highest average gross income among others.

# Finding mode of payment according to cities.

```python
pd.crosstab(df['City'],df['Payment'])
```

```python
pd.crosstab(df['City'],df['Payment']).plot(kind='bar',figsize=(13,5))
plt.xticks(rotation=70,fontsize=10)
plt.show()
```

*  Mandalay -> Most of the customers used Ewallet.
*  Naypyitaw -> Most of the customers used cash.
*  Yangon -> Most of the customers used Ewallet

# Finding Tax values according to product categories.

```python
df.pivot_table(index='Product line',values='Tax 5%',aggfunc='mean').sort_values(by='Tax 5%',ascending =False)
```

```python
df.pivot_table(index='Product line',values='Tax 5%',aggfunc='mean').plot(kind='line',color='purple')
plt.xticks(rotation=80)
plt.show()
```

As we can visualize from the above chart on an average products of Home and Lifestyle has highest tax average and Fashion accessories has lowest tax average on its products.

```python
df.pivot_table(index='Product line',values='Tax 5%',aggfunc='max').sort_values(by='Tax 5%',ascending =False)
```

```python
df.pivot_table(index='Product line',values='Tax 5%',aggfunc='max').plot(kind='line',color='orange')
plt.xticks(rotation=80)
plt.show()
```

As we can see from the line chart Fashion accessories has the maximum tax on products.
